### Directed Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 44

n = ['Vincent', 'Ninya', 'Arton', 'Kassandra', 'Andre']
e = [('Vincent', 'Arton'), ('Arton', 'Kassandra'), ('Kassandra', 'Andre'), ('Andre', 'Ninya')]
color = ['darkseagreen', 'pink', 'tomato', 'gray', 'royalblue']
reto = nx.DiGraph()
reto.add_nodes_from(n)
reto.add_edges_from(e)
pos = nx.spring_layout(reto, seed=seed)

nx.draw(reto, with_labels = True, pos=pos, node_color = color)
```

### Complete Graph

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 10

n = ['Niles', 'Nin', 'Marty', 'Aaron', 'Ian']
color = ['darkseagreen', 'pink', 'tomato', 'gray', 'royalblue']
e = [('Niles', 'Nin'), ('Niles', 'Marty'), ('Niles', 'Aaron'), ('Niles', 'Ian'), ('Marty', 'Nin'), ('Marty', 'Aaron'), ('Aaron', 'Ian')]


friends = nx.Graph()
friends.add_nodes_from(n)
friends.add_edges_from(e)
pos = nx.spring_layout(friends, seed=seed)

nx.draw(friends, with_labels = True, pos=pos, node_color = color)
```

A good way to visualize complete graphs would be to think of a friend group because there is no way to be double friends with someone (Best Friends not counted as parallel edges) which satisfies the requirements to be a complete graph.

### Bipartite Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

color = ['gold', 'pink', 'tomato', 'darkseagreen', 'red']
fo = nx.Graph()
fo.add_nodes_from(['Arton', 'Ella', 'Chynna'], bipartite = 0)
fo.add_nodes_from(['Niles', 'Jess'], bipartite = 1)
fo.add_edges_from([('Niles', 'Arton'), ('Niles', 'Ella'), ('Niles', 'Chynna'), ('Jess', 'Arton'), ('Jess', 'Ella'), ('Jess', 'Chynna')])

top = nx.bipartite.sets(fo)[0]
pos = nx.bipartite_layout(fo, top)

nx.draw(fo, with_labels = True, pos=pos, node_color = color)
```

My example of a Bipartite graph comes from a true experience of still having lots of common friends with someone I cut off a year ago. It shows the connection between the friends(Chynna, Ella, Arton) and the lack of connection from each other(Niles and Jess).

## Travelling in a Graph

#### Walk

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 44

n = ['Bunzel', 'SMED', 'PE', 'LRC', 'SAFAD', 'MR']
color = ['gold', 'lightgreen', 'royalblue', 'silver', 'red', 'tan']
e = [('Bunzel', 'SMED'), ('SMED', 'PE'), ('PE', 'LRC'), ('LRC', 'MR'), ('MR', 'SAFAD') ]

how_to_get_to_safad_on_shuttle = nx.DiGraph()
how_to_get_to_safad_on_shuttle.add_nodes_from(n)
how_to_get_to_safad_on_shuttle.add_edges_from(e)

pos = nx.spring_layout(how_to_get_to_safad_on_shuttle, seed=seed)
nx.draw(how_to_get_to_safad_on_shuttle, with_labels = True, pos=pos, node_color=color)
```

The example I used for a walk is the route to get to SAFAD from Bunzel building, it satisfied the minimum requirements to be a walk.

Bunzel -> SMED -> PE -> LRC -> MR -> SAFAD

It is also possible to skip LRC giving the walk : 
Bunzel -> SMED -> PE -> MR -> SAFAD

Window -> LR -> Kitchen -> DR -> LR -> Bedroom -> Bathroom -> LR -> Grarage

#### Closed Walk

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 1

n = ['Bunzel', 'SMED', 'PE', 'LRC', 'SAFAD', 'MR']
color = ['gold', 'lightgreen', 'royalblue', 'silver', 'red', 'tan']
e = [('Bunzel', 'SMED'), ('SMED', 'PE'), ('PE', 'LRC'), ('LRC', 'MR'), ('MR', 'SAFAD'), ('SAFAD', 'Bunzel'), ('PE', 'MR')]

shuttle_route = nx.DiGraph()
shuttle_route.add_nodes_from(n)
shuttle_route.add_edges_from(e)

pos = nx.spring_layout(shuttle_route, seed=seed)
nx.draw(shuttle_route, with_labels = True, pos=pos, node_color=color)
```

This closed walk has two possible ways : 

Bunzel -> SMED -> PE -> MR -> SAFAD -> Bunzel

or 

Bunzel -> SMED -> PE -> LRC -> MR -> SAFAD -> Bunzel

#### Circuit

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 1

hair_full_cycle = nx.DiGraph()
hair_full_cycle.add_nodes_from(['Grow out Hair', 'Hair too long', 'Get a Haircut', 'Regret', 'Perfect'])
hair_full_cycle.add_edges_from([('Perfect', 'Hair too long'), ('Hair too long', 'Get a Haircut'), ('Get a Haircut', 'Regret'), ('Regret','Grow out Hair'), ('Grow out Hair','Perfect')])

pos = nx.spring_layout(hair_full_cycle, seed=seed)
nx.draw(hair_full_cycle, with_labels = True, pos=pos, node_color = 'grey')
```

This is the full representation of the haircut cycle I believe everyone goes through at least a few times in their lifetime.

## Euler's Circuit

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 1
n = [1, 2, 3, 4, 5]
e = [(1, 2), (2, 3), (3, 1),(1, 4), (4, 5), (5, 1)]

G = nx.DiGraph()
G.add_nodes_from(n)
G.add_edges_from(e)

pos = nx.spring_layout(G)
nx.draw(G, with_labels = True, pos=pos, node_color = 'grey')
```

This Eulers Circuit goes like :

1 -> 2 -> 3 -> 1 -> 4 -> 5 -> 1

which follows every requirement of a Euler's circuit which are
1. Uses every vertex at least once; and
2. Uses every edge Exactly once.

## Hamiltonian Circuits

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 12
n = 6
m = 10

G = nx.gnm_random_graph(n, m, seed=seed)
pos = nx.circular_layout(G)
nx.draw(G, with_labels = True, pos=pos, node_color = 'grey')
```

This is a randomly generated graph which allows for the hamiltonian circuit : 

0 -> 2 -> 3 -> 5 -> 4 -> 1 -> 0

Totalling with 6 Edges and 6 Vertices used while using each exactly once.
